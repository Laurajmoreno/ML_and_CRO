## Functions

### Number of interactions per label in feature

```python
def interactions_per_label(df,feature):
    interactions_per_label=df.groupby(feature).size()
    print(interactions_per_label)
    return pd.DataFrame(interactions_per_label)
```

### Number of conversions per label in feature

```python
def conversions_per_label(df,feature):
    conversions_per_label=df.groupby(feature)['ga:productAddsToCart_transf'].sum()
    print(conversions_per_label)
    return pd.DataFrame(conversions_per_label)
```

### Target prevalence

```python
def calc_prevalence(y):
    return (sum(y)/len(y))*100

calc_prevalence(df['ga:productAddsToCart_transf'])
```

### Target prevalence per label

```python
def prevalence_per_label(df,feature):
    prevalence_per_label=df.groupby(feature).apply(lambda x: calc_prevalence(x['ga:productAddsToCart_transf']))
    print(prevalence_per_label)
    return pd.DataFrame(prevalence_per_label)
```

### Exploratory Analysis

#### Month

La variable 'dateTime_month' ha sido codificada con valores entre 1 y 12, en el que el 1 corresponde al mes de Enero y el 12 al mes de Diciembre.

##### Interactions per month

De acuerdo a los siguientes resultados, los meses que recogen mayor n√∫mero de interacciones son noviembre y enero:

```python
interactions_month=interactions_per_label(df,df['dateTime_month'])
interactions_month.plot()
plt.axhline(y=interactions_month[0].mean(), c= "red", label='mean')
plt.axhline(y=interactions_month[0].median(), c= "green", label='median')
plt.legend()
```

Debemos tener en cuenta que en nuestro dataset se est√°n contemplando los a√±os 2019, 2020 y 2021 y no  hay la misma representatividad para todos los meses de a√±o. Concretamente est√° es la relaci√≥n de interacciones para cada combinaci√≥n MES/A√ëO recogida en el conjunto de datos:


```python
interactions_evol=df.groupby(['dateTime_year','dateTime_month']).size().reset_index().sort_values(by=['dateTime_year','dateTime_month'])
interactions_evol
```

En general, se observa una tendencia creciente mes a mes del n√∫mero de interacciones fruto posiblemente de la aplicaci√≥n de acciones de marketing y, como consecuencia, del crecimiento que est√° experimentando el establecimiento. Caben destacar los picos de:
- Marzo, Abril y Mayo 2020 como consecuencia del confinamiento.
- Noviembre 2020 posiblemente en respuesta a las ofertas y promociones de blackfriday
- Enero 2021 posiblemente fruto del temporal Filomena y las rebajas. 

```python
interactions_evol['month-year']=interactions_evol['dateTime_month'].astype('str')+'-'+(interactions_evol['dateTime_year'].astype('str').str[-2:])
plt.figure(figsize=(15,5))
plt.plot(interactions_evol['month-year'],interactions_evol[0], c='blue', label='interactions')
plt.axhline(y=interactions_evol[0].mean(), c= "red", label='mean')
plt.axhline(y=interactions_evol[0].median(), c= "green", label='median')
plt.title('Interactions evolution over 2019,2020 and 2021')
plt.legend();
```

Al evaluar el detalle de cada uno de los a√±os para evitar la distorsi√≥n que producen las diferentes escalas, se observa que, en 2019, Noviembre tambi√©n fue el mes del a√±o con mayor volumen de interacciones. Asimismo, advertimos caidas de las interacciones en junio y agosto en ambos a√±os:

```python
interactions_2019= interactions_evol[interactions_evol['dateTime_year']==2019]
interactions_2020= interactions_evol[interactions_evol['dateTime_year']==2020]

fig, ax = plt.subplots(2,1,sharex=True)
fig.set_size_inches(15,5)
ax[0].plot(interactions_2019['dateTime_month'],interactions_2019[0], c='b', label='interactions 2019')
ax[0].axhline(y=interactions_2019[0].mean(), c= "red", label='mean')
ax[0].axhline(y=interactions_2019[0].median(), c= "green", label='median')
ax[0].set_title('Interactions per month 2019')
ax[0].legend()
ax[1].plot(interactions_2020['dateTime_month'],interactions_2020[0], c='b', label='interactions 2020')
ax[1].axhline(y=interactions_2020[0].mean(), c= "red", label='mean')
ax[1].axhline(y=interactions_2020[0].median(), c= "green", label='median')
ax[1].set_title('Interactions per month 2020')
ax[1].legend()
```

##### Comparative plot

```python
def day_of_week(s):
    days_of_week=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    for i, day in enumerate(days_of_week):
        if s==i:
            return day

interactions_dayofweek= interactions_dayofweek.reset_index()
conversions_dayofweek=conversions_dayofweek.reset_index()
prevalence_dayofweek= prevalence_dayofweek.reset_index()
        
interactions_dayofweek['day_of_week']=prevalence_dayofweek['dateTime_dayofweek'].apply(day_of_week)
conversions_dayofweek['day_of_week']=prevalence_dayofweek['dateTime_dayofweek'].apply(day_of_week)
prevalence_dayofweek['day_of_week']=prevalence_dayofweek['dateTime_dayofweek'].apply(day_of_week)

fig, ax = plt.subplots(3,1,sharex=True)
fig.set_size_inches(10,8)
ax[0].plot(interactions_dayofweek['day_of_week'],interactions_dayofweek[0], c='b', label='interactions')
ax[0].axhline(y=interactions_dayofweek[0].mean(), c= "red", label='mean')
ax[0].axhline(y=interactions_dayofweek[0].median(), c= "green", label='median')
ax[0].legend()
ax[0].set_title('Interactions')
ax[1].plot(conversions_dayofweek['day_of_week'],conversions_dayofweek['ga:productAddsToCart_transf'], c='b', label='conversions')
ax[1].axhline(y=conversions_dayofweek['ga:productAddsToCart_transf'].mean(), c= "red", label='mean')
ax[1].axhline(y=conversions_dayofweek['ga:productAddsToCart_transf'].median(), c= "green", label='median')
ax[1].legend()
ax[1].set_title('Conversions')
ax[2].plot(prevalence_dayofweek['day_of_week'],prevalence_dayofweek[0], c='b', label='prevalence')
ax[2].axhline(y=prevalence_dayofweek[0].mean(), c= "red", label='mean')
ax[2].axhline(y=prevalence_dayofweek[0].median(), c= "green", label='median')
ax[2].legend()
ax[2].set_title('Target prevalence');
```

El domingo es el mejor d√≠a en t√©rminos de interacciones y conversiones. Ahora bien, la prevalencia ese d√≠a nos indica que no crecen las conversiones en la misma proporci√≥n que las interacciones. Es posible que los usuarios, dada una mayor disponibilidad de tiempo, pasen m√°s tiempo explorando y comparando en la web, incrementando as√≠ las interacciones y penalizando la prevalencia. 

En la otra cara de la moneda, tenemos los jueves y viernes que cuentan uno de los mejores ratios de prevalencia (por delante de los domingos), dada unas conversiones altas en relaci√≥n a sus interacciones. (OJO: El Blackfriday que concentra un gran volumen de ventas puede estar sesgando estos resultados)

Los lunes, por su parte, muestran una relaci√≥n m√°s coherente entre interacciones, conversiones y prevalencia.
