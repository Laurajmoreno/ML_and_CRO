## Functions

### Number of interactions per label in feature

```python
def interactions_per_label(df,feature):
    interactions_per_label=df.groupby(feature).size()
    print(interactions_per_label)
    return pd.DataFrame(interactions_per_label)
```

### Number of conversions per label in feature

```python
def conversions_per_label(df,feature):
    conversions_per_label=df.groupby(feature)['ga:productAddsToCart_transf'].sum()
    print(conversions_per_label)
    return pd.DataFrame(conversions_per_label)
```

### Target prevalence

```python
def calc_prevalence(y):
    return (sum(y)/len(y))*100

calc_prevalence(df['ga:productAddsToCart_transf'])
```

```python
def evaluate(X_test, y_test, model):
    X_test=sm.add_constant(X_test)
    probabilities=model.predict(X_test)
    predictions=np.where(probabilities<0.5,0,1)
    print("ACCURACY: " + str(accuracy_score(y_test, predictions)))
    print("PRECISION: " + str(precision_score(y_test, predictions)))
    print("RECALL: " + str(recall_score(y_test, predictions)))
    print("F1 SCORE: " + str(f1_score(y_test, predictions)))
    print("AUC: " + str(roc_auc_score(y_test,probabilities)))
    tpr, fpr, thresolds = roc_curve(y_test, probabilities)
    plt.plot(tpr, fpr)
    plt.xlabel('fpr')
    plt.ylabel('tpr')
```

### Model Performance Report

```python
def report(X_test,y_test, model):
    X_test=sm.add_constant(X_test)
    probabilities=model.predict(X_test)
    predictions=np.where(probabilities<0.5,0,1)
    confusion_matrix=pd.crosstab(y_test.ravel(),predictions,rownames=['Actual values'],colnames=['Predictions'])
    print(classification_report(y_test, predictions))
    return confusion_matrix
```

```python
# DateTime as cyclical features:

df['month_sin']=np.sin((df.dateTime_month-1)*(2.*np.pi/12))
df['month_cos']=np.cos((df.dateTime_month-1)*(2.*np.pi/12))
# I substract minus 1 to 'df.dateTime_month' because its values are coded as 1 to 12 instead of 0 to 11

df['dayofweek_sin']=np.sin(df.dateTime_dayofweek*(2.*np.pi/7))
df['dayofweek_cos']=np.cos(df.dateTime_dayofweek*(2.*np.pi/7))

df['hour_sin']=np.sin(df.dateTime_hour*(2.*np.pi/24))
df['hour_cos']=np.cos(df.dateTime_hour*(2.*np.pi/24))

df.info()
```

### Exploratory Analysis

#### Month

La variable 'dateTime_month' ha sido codificada con valores entre 1 y 12, en el que el 1 corresponde al mes de Enero y el 12 al mes de Diciembre.

##### Interactions per month

De acuerdo a los siguientes resultados, los meses que recogen mayor número de interacciones son noviembre y enero:

```python
interactions_month=interactions_per_label(df,df['dateTime_month'])
interactions_month.plot()
plt.axhline(y=interactions_month[0].mean(), c= "red", label='mean')
plt.axhline(y=interactions_month[0].median(), c= "green", label='median')
plt.legend()
```

Debemos tener en cuenta que en nuestro dataset se están contemplando los años 2019, 2020 y 2021 y no  hay la misma representatividad para todos los meses de año. Concretamente está es la relación de interacciones para cada combinación MES/AÑO recogida en el conjunto de datos:


```python
interactions_evol=df.groupby(['dateTime_year','dateTime_month']).size().reset_index().sort_values(by=['dateTime_year','dateTime_month'])
interactions_evol
```

En general, se observa una tendencia creciente mes a mes del número de interacciones fruto posiblemente de la aplicación de acciones de marketing y, como consecuencia, del crecimiento que está experimentando el establecimiento. Caben destacar los picos de:
- Marzo, Abril y Mayo 2020 como consecuencia del confinamiento.
- Noviembre 2020 posiblemente en respuesta a las ofertas y promociones de blackfriday
- Enero 2021 posiblemente fruto del temporal Filomena y las rebajas. 

```python
interactions_evol['month-year']=interactions_evol['dateTime_month'].astype('str')+'-'+(interactions_evol['dateTime_year'].astype('str').str[-2:])
plt.figure(figsize=(15,5))
plt.plot(interactions_evol['month-year'],interactions_evol[0], c='blue', label='interactions')
plt.axhline(y=interactions_evol[0].mean(), c= "red", label='mean')
plt.axhline(y=interactions_evol[0].median(), c= "green", label='median')
plt.title('Interactions evolution over 2019,2020 and 2021')
plt.legend();
```

Al evaluar el detalle de cada uno de los años para evitar la distorsión que producen las diferentes escalas, se observa que, en 2019, Noviembre también fue el mes del año con mayor volumen de interacciones. Asimismo, advertimos caidas de las interacciones en junio y agosto en ambos años:

```python
interactions_2019= interactions_evol[interactions_evol['dateTime_year']==2019]
interactions_2020= interactions_evol[interactions_evol['dateTime_year']==2020]

fig, ax = plt.subplots(2,1,sharex=True)
fig.set_size_inches(15,5)
ax[0].plot(interactions_2019['dateTime_month'],interactions_2019[0], c='b', label='interactions 2019')
ax[0].axhline(y=interactions_2019[0].mean(), c= "red", label='mean')
ax[0].axhline(y=interactions_2019[0].median(), c= "green", label='median')
ax[0].set_title('Interactions per month 2019')
ax[0].legend()
ax[1].plot(interactions_2020['dateTime_month'],interactions_2020[0], c='b', label='interactions 2020')
ax[1].axhline(y=interactions_2020[0].mean(), c= "red", label='mean')
ax[1].axhline(y=interactions_2020[0].median(), c= "green", label='median')
ax[1].set_title('Interactions per month 2020')
ax[1].legend()
```
