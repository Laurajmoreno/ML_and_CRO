### *Feature Engineering*

Ahora mismo, esta variable es del tipo numérico, por lo que la convertiremos a formato *datetime*:

```python
df['ga:dateHourMinute']
```

```python
df['dateTime']=pd.to_datetime(df['ga:dateHourMinute'],format='%Y%m%d%H%M')
df['dateTime']
```

Una vez transformada, extraemos sus distintos componentes para hacer un análisis pormenorizado de cada uno de ellos:

```python
df['dateTime_year']=df['dateTime'].dt.year
df['dateTime_month']=df['dateTime'].dt.month
df['dateTime_dayofweek']=df['dateTime'].dt.dayofweek
df['dateTime_hour']=df['dateTime'].dt.hour
df.info()
```

Asimismo, de cara al modelo, crearemos el seno y el coseno de cada una de ellas, para dotarlas de su carácter cíclico y continuo, de modo que el menor (ejemplo: enero-1) de sus valores aparezca junto al mayor (ejemplo: diciembre-12):

```python
# DateTime as cyclical features:

df['month_sin']=np.sin((df.dateTime_month-1)*(2.*np.pi/12))
df['month_cos']=np.cos((df.dateTime_month-1)*(2.*np.pi/12))
# I substract minus 1 to 'df.dateTime_month' because its values are coded as 1 to 12 instead of 0 to 11

df['dayofweek_sin']=np.sin(df.dateTime_dayofweek*(2.*np.pi/7))
df['dayofweek_cos']=np.cos(df.dateTime_dayofweek*(2.*np.pi/7))

df['hour_sin']=np.sin(df.dateTime_hour*(2.*np.pi/24))
df['hour_cos']=np.cos(df.dateTime_hour*(2.*np.pi/24))

df.info()
```

### 'Source'

#### *Feature Engineering*

Si atendemos a la fuente, observamos que el top 30 de las páginas representan el 99,7% de las observaciones:

```python
n=30
print(df['Source'].value_counts(normalize=True).head(n).sum())
df['Source'].value_counts().head(n)
```

Algunas de estas páginas, como facebook o instagram, aparecen repetidas bajo distintas formas. Las agruparemos para facilitar el análisis.

Cabe señalar que *linktr.ee* es el agregador de enlaces que utiliza Galileo61 en su perfil de instagram para direccionar a los usuarios a su web.

```python
def source_compiler(x):
    
    insta_words= ['instagram','ig','linktr.ee']
    for seq in insta_words:
        if seq in x.lower():
            return 'instagram'
    if 'facebook' in x.lower():
        return 'facebook'
    if 'youtube' in x.lower():
        return 'youtube'  
    if 'vogue' in x.lower():
        return 'vogue'
    if 'yahoo' in x.lower():
        return 'yahoo'
    else:
        return x

df['Source']=df['Source'].apply(source_compiler)
df['Source'].value_counts()
```

Una vez compiladas las referencias repetidas obtenemos que el **99,4%** de las observaciones proceden del **top 10** de las páginas. Nos quedaremos con ellas y crearemos una nueva categoría que englobe al resto:

### 'Medium'

Esta variable identifica la tipología de tráfico.

#### Feature Engineering

En este caso, tenemos un número mucho más limitado de etiquetas, también con algunas repeticiones:

```python
df['Medium'].value_counts()
```

En este caso, compilaremos las etiquetas de acuerdo a lo siguiente:
- ***CPC*** se repite en mayúsculas y minúsculas. Ambos corresponden a un tipo de tráfico de pago para distintos fuentes. Dado que ya contamos con esa información en *Source*, los agruparemos.
- ***Social*** corresponde a IGShopping. Lo designaremos así para diferenciarlo del tráfico que procede de Instagram como referral. 
- ***facebook, facebook feed y trafico*** corresponden a distintos tipos de tráfico de facebook. Dado que en *Source* lo agregamos, aquí intentaremos ser más precisos. En el caso de *facebook*, vemos que corresponde a la antigua fuente *facebook ads*, por lo que lo convertiremos en este valor. Asimismo, conservaremos *facebook feed* y cambiaremos el nombre de *trafico* por *facebook trafico* para evitar confusiones futuras.

```python
def medium_compiler(x):
    
    if 'cpc' in x.lower():
        return 'cpc'
    if x=='Social':
        return 'IGShopping'  
    if x=='facebook':
        return 'facebook ads'
    if x=='trafico':
        return 'facebook trafico'
    else:
        return x

df['Medium']=df['Medium'].apply(medium_compiler)
df['Medium'].value_counts()
```
